```python
expr = x*x + x*y+x*y+y*y
res = expr.subs({x:3,y:5})
res
```

```python
from sympy import Symbol, solve
expr = x**2 + 5*x + 4
solve(expr, dict=True)
```

```python
expr1 = 2*x + 3*y - 6
expr2 = 3*x + 2*y - 12
#左辺を０にする

solve((expr1, expr2), dict=True)
#定義した式をタプルにして、solve関数に入れる
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 6, 0.5)

y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```
